Matplotlib
   
    Matriz de Correlacion con un mapa de color

- Mostrar Imagen y arreglos de visualizacion
- plt.imshow() es una funcion de matplotlib que se utiliza para mostrar imagenes 2D, visualizar matrices 2D o mostrar imagenes en varios formatos
- El uso de imshow para un mapa de calor: Un mapa de Calor es una visualizacion para que una matriz de correlacion, que nos dara una idea de como cada variable esta correlacionada con otras variables.
- Aqui crearemos un mapa de calor para visualizar una matriz de correlacion y utilizaremos un mapa de colores para mostrar esta relacion visualmente.
- Pasa la matriz de correlaci칩n a imshow para visualizar el mapa de calor.
---
- Uso de Datos Aleatorios
---

```python
# Librerias
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Crear una matriz de correlaci칩n de muestra
matriz_correlacion = np.array ([[1.0,0.8,0.3,-0.2],
                                [0.8,1.0,0.5,0.1],
                                [0.3,0.5,0.1,-0.4],
                                [-0.2,0.1,-0.4,1.0]])
# Crear un mapa de calor para la matriz de correlaci칩n
plt.figure(figsize=(8,6)) # tama침o de la figura o diagrama
# Parametros
# matriz_correlacion = contiene los datos que se desea visualizar
# cmap = 'coolwarm' determina el mapa de colores que se utilizara para representar el mapa de calor
# vmin = -1  Especifica el valor minimo en el mapa de colores
# vmax = 2  Especifica el valor maximo en el mapa de colores
# aspect = 'auto'  Controla el aspecto del mapa de calor
# origin = 'upper' Define la ubicacion del origen del eje en la esquina superior
plt.imshow(matriz_correlacion,cmap='coolwarm',vmin = -1, vmax = 1, aspect = 'auto',origin ='upper')
# Agregar una barra de colores
cbar = plt.colorbar()
# Este metodo es para agregar una etiqueta a la barra de color ('colorbar')
# rotation = 270 indica el angulo de rotacion del texto en la etiqueta - En este caso para que este orientada verticalmente
# labelpad = 20  controla la distancia entre la barra de color y la etiqueta permitiendo ajustar el espacio para mejorar la lectura
cbar.set_label('Correlacion', rotation = 270, labelpad = 20)

# Agregar Etiquetas y Titulos
plt.title ('Correlacion de Mapa de Calor')
plt.xticks(range(len(matriz_correlacion)),['variable_1','variable_2','variable_3','variable_4'])
plt.yticks(range(len(matriz_correlacion)),['variable_1','variable_2','variable_3','variable_4'])
# Mostrar el Grafico
plt.show()
```
