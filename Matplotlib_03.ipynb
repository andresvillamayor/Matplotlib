Ejemplo con Datos Aleatorios

- Categoria de Gastos y Comparación , por persona


```python
# Librerias
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Categoria de Gastos en Dolares (USD)
# casa(gastos) = los gastos que una persona puede tener en una casa o departamento
# transporte (gastos) = los gastos de movilidad, compra de combustible, taxi, uber, transporte
# alimento (gastos) = los gastos de alimento que uno consume
# salidas (gastos) = gastos de salidas a tomar un cafe , cine, teatro , cenas
# servicio (gastos) = gastos de agua, luz, electricidad.
categoria_gastos = ['casa','transporte','alimento','salidas','servicios']
```

```python
# Gastos Mensuales  en USD que tienen (Juan,Jose y Lorena) mensualmente son :
gastos_juan = [300,100,250,120,80]
gastos_jose = [250,110,230,100,90]
gastos_lorena = [280,115,280,80,120]

# Generar un array para el eje de las x
x = np.arange(len(categoria_gastos))

# Establecer el ancho de las barras
ancho_barra = 0.2

# Creamos la barra de los gastos mensuales en Dolares de Juan
# plt.bar  = crea el grafico
# x - ancho_barra = es la ubicacion en el eje x
# gastos_juan = son los valores de los gastos
# label = es la etiqueta que le asigo a las barras que sería Juan
# color = es el color asignado a la barra de juan
plt.bar(x - ancho_barra,gastos_juan, width = ancho_barra, label ='Juan', color ='skyblue')

# Creamos la barra de los gastos mensuales de jose
plt.bar(x, gastos_jose, width = ancho_barra, label ='Jose', color ='lightcoral')

# Creamos la barra de los gastos mensuales de Lorena
plt.bar(x + ancho_barra,gastos_lorena, width = ancho_barra, label = 'Lorena', color = 'lightgreen')

# Agregamos al Gráfico las etiqetas, titulos y legendas
plt.xlabel('Categoria de Gastos')
plt.ylabel ('Gastos Mensuales en USD')
plt.title ('Comparación de Gastos Mensuales en USD')
# para personalizar y configurar las etiquetas en el eje x del grafico
plt.xticks (x,categoria_gastos)
plt.legend()
# Mostrar el Grafico
plt.show()
```

Bar Plots con Iris

---
Diccionario de Datos

- sepal_length = Longitud del sépalo en centímetros
- sepal_width = ancho del sépalo en centímetros
- petal_length = longitud del pétalo en centímetros
- petal_width = ancho del pétalo en centimetros
- species = especie de iris (Setosa, Versicolor o Virginica)
---

```python
# Librerias
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from sklearn.datasets import load_iris
```

```python
# Cargar el conjunto de datos Iris
iris = load_iris()
data = iris['data']
target = iris['target']
target_names = iris['target_names']
```

```python
# Contar la cantidad de muestras por especie
species_count = {species: sum(target == i) for i, species in enumerate(target_names)}
```

```python
# Crear el gráfico de barras
plt.figure(figsize=(5, 5))
sns.barplot(x=list(species_count.keys()), y=list(species_count.values()))
plt.xlabel('Especies de Iris')
plt.ylabel('Cantidad de Muestras')
plt.title('Cantidad de Muestras por Especie de Iris')
plt.show()
```

---
* En este código, se utiliza **bar_plot.text()** para agregar etiquetas de texto encima de cada barra en el gráfico de barras. Estas etiquetas muestran la cantidad de muestras para cada especie de flor en el conjunto de datos Iris.

```python
# Contar la cantidad de muestras por especie
species_count = {species: sum(target == i) for i, species in enumerate(target_names)}

# Crear el gráfico de barras con etiquetas de cantidad de muestras
plt.figure(figsize=(5, 5))
bar_plot = sns.barplot(x=list(species_count.keys()), y=list(species_count.values()))

# Mostrar las cantidades de muestras encima de las barras
for index, value in enumerate(species_count.values()):
    bar_plot.text(index, value + 1, str(value), ha='center', color='red', fontweight='bold')

plt.xlabel('Especies de Iris')
plt.ylabel('Cantidad de Muestras')
plt.title('Cantidad de Muestras por Especie de Iris')
plt.show()
```
