Matplotlib
- Scatter Plot (Diagrama de Dispersión)
  - Los diagramas de dispersión muestran puntos de datos individuales como puntos en un plano bidimensional.
  - Se utilizan para explorar relaciones entre dos variables numéricas, representado un valor en el eje $x$ y otro en el eje $y$.

- El objetivo principal de un scatter plot es visualizar la relación o patrón entre dos variables. Puedes observar la distribución de los puntos en el gráfico para identificar tendencias, patrones, correlaciones o cualquier otra estructura que pueda existir entre las dos variables representadas en los ejes.

```python
# Librerias
import numpy as np
import matplotlib.pyplot as plt
```

---
Ejemplo con Datos Aleatorios

```python
# Ejmplo 1
# Ejemplo de un gráfico de dispersión que analiza la relacion entre número de clientes, los ingresos y el tamaño de la tienda.

# tiendas
tiendas = ['tienda_A','tienda_B', 'tienda_C','tienda_D','tienda_E']
# cantidad de clientes
clientes = [120,90,150,85,220]
# ingresos en las tiendas en USD
ingresos = [20000,15000,32000,16000,10000]
# tamaño de la tienda por metro cuadrados
tienda_tamaño =[30,15,42,23,10]

# Se ajusta una escala  para el tamñano de la tienda asi determinar el tamño de los puntos
# Esto puede utilizarse  si uno necesita o no
puntos_tamaño = [size * 100 for size in tienda_tamaño]

# Creamos un gráfico de dispersion con diferentes tamaños de los puntos
plt.figure(figsize=(10,8))
# parametros
# s = especifica el tamaño de los puntos en el grafico
# c = define el color de los puntos de los graficos
# alpha = transparencia de los puntos , es util si hay superposición de puntos
# edgecolors = determina el color de los bordes de los puntos
plt.scatter(clientes,ingresos, s = puntos_tamaño, c = 'skyblue', alpha = 0.7, edgecolors = 'b')

# Agegamos etiquetas, titulos y legendas
plt.xlabel ('Numero de Clientes')
plt.ylabel ('Ingresos en USD')
plt.title  ('Relación entre clintes, ingresos y tamaño de la tienda')

# Mostrar el gráfico
# Aplicamos que tenga grilla el gráfico
plt.grid(True)
plt.show()
```

Ejemplo con Datos de Iris
  - Utilización de Seaborn para mejorar el gráfico

```python
# Librerias
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import load_iris
import pandas as pd
```

```python
# Cargar el conjunto de datos Iris
iris = load_iris()
iris_data = iris.data
iris_target = iris.target

# Crear un DataFrame de pandas para facilitar el trabajo con los datos
iris_df = pd.DataFrame(iris_data, columns=iris.feature_names)
iris_df['Target'] = iris_target

# Establecer el estilo de seaborn para gráficos más atractivos
sns.set(style="whitegrid")

# Crear un scatter plot utilizando seaborn
sns.scatterplot(x='sepal length (cm)', y='sepal width (cm)', hue='Target', data=iris_df, palette='viridis', legend='full')

# Añadir etiquetas y título
plt.xlabel('Longitud del Sépalo (cm)')
plt.ylabel('Ancho del Sépalo (cm)')
plt.title('Scatter Plot de Iris: Sépalo')

# Mostrar el gráfico
plt.show()
```
