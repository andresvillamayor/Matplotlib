```python
# Librerias
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
```

Cargar los datos de Iris

```python
iris = load_iris()
data = iris.data
target = iris.target
feature_names = iris.feature_names
```

```python
# Creando el gráfico con Data de Iris
# Estableciendo el tamaño del gráfico
plt.figure(figsize=(10, 6))
# Recorriendo todas las características
for i in range(data.shape[1]):
    # Graficando cada característica
    # feature_names contiene la lista de nombres, facilitando la comprensión y referencia
    plt.plot(data[:, i], label=feature_names[i])
# Agregar títulos y etiquetas
plt.xlabel('Muestras')
plt.ylabel('Valor')
plt.title('Gráfico de Múltiples Líneas - Conjunto de datos Iris')
# Mostrar leyenda con los nombres de las características
plt.legend()
# Mostrar el gráfico
plt.show()
```

---
Ejemplo con datos aleatorios
- Meses
- Ingresos del Producto A
- Ingresos del Produto B

```python
# Ejemplo de datos aleatorios
# Ingresos mensuales para dos productos, cada uno representado en miles de dólares
meses = ['Enero','Febrero','Marzo','Abril','Mayo','Junio']
ingresos_producto_a = [45,55,65,45,58,69]
ingresos_producto_b = [35,28,15,40,29,11]
# Parámetros a ser utilizados
# marker ='o' es para representar en circulos
# marker ='s' es para representar en cuadrados
# linestyle = '-' representa una línea sólida (predeterminada)
# linestyle = '--' representa una línea discontinua

# Crear el gráfico de línea para Producto A con línea azul y marcadores circulares
plt.plot(meses,ingresos_producto_a,marker='o',linestyle='-', color = 'blue', label ='Producto A')
# Crear el gráfico de línea para Producto B con línea roja y marcadores cuadrados
plt.plot(meses,ingresos_producto_b,marker='s',linestyle='--', color = 'red', label ='Producto B')

# Agregar títulos y etiquetas
plt.xlabel ('Meses')
plt.ylabel ('Ingresos Mensuales (en USD-1000)')
plt.title ('Comparacion de Ingresos Mensuales entre los Productos')

# Mostrar Leyenda para diferenciar los productos
plt.legend()
# Mostrar el Gráfico
plt.show()
```
