Diccionario de Datos


*   sepal_length = Longitud del sépalo en centímetros
*   sepal_width = ancho del sépalo en centímetros
*   petal_length = longitud del pétalo en centímetros
*   petal_width = ancho del pétalo en centimetros
*   species = especie de iris (Setosa, Versicolor o Virginica)



```python
# Librerias
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
```

Cargar los datos de Iris


```python
iris = load_iris()
data = iris.data
target = iris.target
```

```python
print('Datos de las Flores de Iris')
print (iris.data[:5])
```

----
Simple Line Plots
*  Con datos de Iris- Ejemplo 1

```python
# Obtener la longitud del sépalo para todas las muestras
sepal_length = iris.data[:, 0]  # Índice 0 representa la longitud del sépalo

# Crear un gráfico de línea simple
plt.plot(sepal_length)
plt.xlabel('Muestra')
plt.ylabel('Longitud del sépalo (cm)')
plt.title('Longitud del sépalo en el conjunto de datos Iris')
plt.show()
```

Datos de Muestra que representa el tráfico mensual de un sitio web (en miles)

```python
meses = ['Enero','Febrero','Marzo','Abril','Mayo','Junio']
trafico= [150,170,120,100,200,250]
# Crear un Grafico de linea Simple
plt.plot(meses,trafico)
plt.xlabel('Meses')
plt.ylabel('Trafico de la web (Miles)')
plt.title('Datos del trafico de la web por mes')
plt.show()
```

Utilizando el parámetro marker para marcar los puntos


1.   Linestyle = cambiar el estilo de la linea
2.   grid = añadir la cuadricula al grafico



```python
# Ejemplo 1 ( datos con Iris)
sepal_length = iris.data[:,0] # Se accede al primera columna de datos
# Crear el grafico
plt.plot(sepal_length,marker='o', linestyle='--', color ='b')  # Cambia el estilo de línea
plt.xlabel('Muestra')
plt.ylabel('Longitud del sépalo (cm)')
plt.title('Longitud del sépalo en el conjunto de datos Iris')
plt.grid(True)  # Añade una cuadrícula al gráfico
plt.show()
```

Datos de Muestra que representa el tráfico mensual de un sitio web (en miles)

```python
# Generar el grafico de transito en la web
plt.plot(meses,trafico,marker ='o',linestyle= '--', color = 'g')
# Agregar los titulos y etiquetas
plt.xlabel('Meses')
plt.ylabel('Trafico de la web (Miles)')
plt.title('Datos del trafico de la web por mes')
plt.grid(True) # Muestra la grilla
# Muestra el Grafico
plt.show()
```
