**Pie Charts**
- Los pie charts o gráficos circulares representas partes de un todo.
- Son adecuados para mostrar la composición de una única variable categórica en términos de porcentajes. Sin embargo, esto no se verá bien cuando hay más de seis categorías, ya que se vuelven confusos; en tales casos, podría preferirse un gráfico de barras horizontal.

**Ejemplo**
- Ejemplo de Gastos por categorias (Datos Aleatorios)
- Ejemplo de Ventas por productos  (Datos Aleatorios)


```python
# Librerias
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Ejemplo de Gastos
gastos = ['Casa','Transporte','Alimento','Salidas','Servicios']
# Los montos de los Gastos estan en Dolares- USD
monto = [190,90,100,80,85]
# Creamos el pie chart
# %1.1f: Esto especifica que se mostrará un número de punto flotante con un dígito entero y un dígito decimal.
# %%: El doble símbolo de porcentaje (%%) se utiliza para representar el signo de porcentaje literal ("%") en la cadena de formato.
plt.pie(monto, labels= gastos,autopct = '%1.1f%%')
# Agregar el Titulo
plt.title ('Gastos del Mes por Categoria')
# Mostrar el Grafico
plt.show()
```

---
También podemos resaltar segmentos particulares utilizando el parámetro explode.

```python
# Ejempo 2
productos =['Electronica','Ropa','Libros','Jugetes']
# Datos de ventas por cada producto en  USD
venta =[4000,3200,2500,4500]
# Usamos Explode para el producto de libros
explode  = (0,0,0.1,0)
# Creamos el pie chart con el explode y le agregamos sombra al gráfico
# Colocamos el tamaño de la figura
plt.figure(figsize=(6,6))
plt.pie (venta,labels= productos, explode = explode, shadow = True,autopct = '%1.1f%%')
# Agregar el titulo
plt.title ('Ventas por Producto')
# Mostrar el Gráfico
plt.show()
```
