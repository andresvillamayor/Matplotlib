Histogramas

* Uso de Datos alatorios de Ejemplo

* Los histogramas se utilizan para visualizar la distribucion de datos continuos o númericos, y que nos ayudan a identificar patrones en los datos.

* En un gráfico de histograma, los datos se agrupan en bins (intervalos), y a la altura de cada barra representa la frecuencia o cantidad de puntos de datos
en ese bin. Toman los limites inferior y superior de los datos dados y los
dividen en los números de bins.

* Pueden usar el comando plt.hist(x) para generar histogramas. A diferencia del
grafico de barra, aqui no necesitas el 'y' , ya que solo representa la frecuencia de un dato continuo. Los bins predeterminado son 10 y se pueden cambiar. Puedes sobreescribir el rango de los bins con el rango de bins que desees. Tambien puedes agregar el 'edgecolor' para las barras.

* junto con el gráfico de histograma, en el mismo gráfico, si deseas agregar alguna línea, como la media o la mediana se puede hacer calculando el valor y pasándolo a plt.axline(calculated_mean, label = desited_label). Esto se puede utilizar con cualquier otro grafico.

* Algunos casos de uso:

 - Analizar la distribución de edades de una población
 - Examinar la distribución de puntuaciones de exámenes en un aula

```python
# Librerias
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Distribucion de una poblacion por edades
# Datos de puntaje de examenes como muestra
exam_score = [68,72,75,80,82,84,86,90,92,95,98,100]

# Creamos un Histograma sin especificar rangos de bin
plt.figure(figsize = (5,5)) # Establece el sitio de la figura
plt.hist(exam_score, color= 'orange', edgecolor ='black')

# Agregar Labels y titulos
plt.xlabel ('Puntaje de Examenes')
plt.ylabel ('Frecuencia')
plt.title  ('Histograma de Puntaje de Examenes - Bins autogenerados')

# Mostrar el Histograma
plt.show()
```

```python
# Ejemplo 2
# Datos de puntajes de examenes - muestra
exam_scores = [68,72,75,80,82,84,86,90,92,95,98,100]
# Rango de bins personalizados
bin_ranges = [60,70,80,90,100]
# Crear un histograma con el rango de bins personalizados
plt.figure(figsize = (5,5)) # Tamaño de la figura
plt.hist(exam_scores,bins = bin_ranges, color = 'orange', edgecolor ='black', alpha = 0.7)

# Agregar Titulos y nombres
plt.xlabel('Puntaje de Examenes')
plt.ylabel('Frecuencia')
plt.title('Histograma Puntaje de Examenes bajo bins')

# Calcula y suma una linea median
median_score = np.median(exam_scores)
plt.axvline(median_score, color = 'red', linestyle = 'dashed', linewidth = 2 ,label = f'Valor de la Mediana:{median_score}')

# Agrega la leyenda
plt.legend()

# Display the plot
plt.grid(True) # Se agrega la cuadricula para una mejor lectura
plt.show()

# En este ejemplo
# Se tiene un conjunto de datos de puntuaciones de exámenes
# Especificamos rangos de contenedores personalizados (bin)
# plt.hist() la funcion se utiliza con bin personalizado estableciendo color , el color del borde y la transparencia (alpha) en las barras
```
